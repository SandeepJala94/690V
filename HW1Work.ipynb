```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import HoverTool
output_notebook()
```

```python
import xlrd
file_location = "C:/Users/Sandeep/Desktop/CollegeCourses/GRAD_SCHOOL/CS_690V_Visual_Analytics/HW/HW1/indicator_m_15-24_employ.xlsx"
workbook = xlrd.open_workbook(file_location)
sheet = workbook.sheet_by_index(0)
```

```python
sheet.cell_value(0, 0)
print(sheet.nrows)
print(sheet.ncols)
```

```python
data = [[sheet.cell_value(r, c) for c in range(sheet.ncols)] for r in range(sheet.nrows)]
```

```python
for c in range(sheet.ncols):
    print(data[0][c]) 
```

```python
years = []
for c in range(1, sheet.ncols):
    years.append(sheet.cell_value(0, c))
    
print(years)
```

```python
for i in range(len(years)):
    years[i] = years[i] - 1991
    
print(years)
```

```python
US_employment_rates = []
for c in range(1, sheet.ncols):
    US_employment_rates.append(sheet.cell_value(170, c))
    
print(US_employment_rates)
print(len(US_employment_rates))
```

```python
hoverTwo = HoverTool(
        tooltips=[
            ("index", "$index"),
            ("years after 1991", "@x"),
            ("US employment rate", "@y"),
        ]
    )

# create a new plot with default tools, using figure
p = figure(plot_width=400, plot_height=400, tools=[hoverTwo], title="US Employment Rates after 1991")

# add a circle renderer with a size, color, and alpha
p.circle(years, US_employment_rates, size=15, line_color="navy", fill_color="orange", fill_alpha=0.5)

show(p) # show the results
```

```python
five_countries = [sheet.cell_value(170, 0), sheet.cell_value(30,0), sheet.cell_value(34,0), sheet.cell_value(76,0), sheet.cell_value(136,0)]

for i in five_countries:
    print(i)
```

```python
# US, Cape Verde, China, Iran, Russia
five_countries_employment_rates = [sheet.cell_value(170, 17), sheet.cell_value(30,17), sheet.cell_value(34,17), sheet.cell_value(76,17), sheet.cell_value(136,17)]

for i in five_countries_employment_rates:
    print(i)
```

```python
hoverThree = HoverTool(
        tooltips=[
            ("Employment Rate", "@top"),
        ]
    )

# Here is a list of categorical values (or factors)
fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']

# Set the x_range to the list of categories above
p = figure(x_range=five_countries, plot_height=250, tools=[hoverThree], title="Employment Rates of Men Ages 15-24 in 5 Countries for 2007")

# Categorical values can also be used as coordinates
p.vbar(x=five_countries, top=five_countries_employment_rates, width=0.9)

# Set some properties to make the plot look better
p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
UK_employment_rates = []
for c in range(1, sheet.ncols):
    UK_employment_rates.append(sheet.cell_value(169, c))
print(UK_employment_rates)
```

```python
hoverFour = HoverTool(
        tooltips=[
            ("index", "$index"),
            ("Year after 1991", "@x"),
            ("Employment Rate", "@y"),
        ]
    )

# create a new plot (with a title) using figure
p = figure(plot_width=400, plot_height=400, tools=[hoverFour], title="My Line Plot")
    
# add a line and circle renderer
p.line(years, UK_employment_rates, line_width=2)
p.circle(years, UK_employment_rates, fill_color="white", size=8)

show(p) # show the results
```
